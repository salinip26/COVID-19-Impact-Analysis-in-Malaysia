```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')  # to get seaborn scatter plot

# read the csv file to extract data

data = pd.read_csv('firstlockdown.csv')
view_count = data['date']
icu = data['icu']


plt.scatter(view_count,icu, s=100, alpha=0.6, edgecolor='black', linewidth=1)
plt.xticks(rotation=90)
plt.title('ICU Patients')
plt.xlabel('Date')
plt.ylabel('ICU')

plt.tight_layout()
plt.show()
```
